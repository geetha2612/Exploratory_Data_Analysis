BIVARIATE ANALYSIS

```python
import seaborn as sns
```

```python
df.isna().sum()
```

```python
drop = df.dropna()
```

```python
drop.isna().sum()
```

```python
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
plt.title('Sales vs PV(r = {0:0.2f})'.format(pearsonr(drop['Sales'],drop['PV'])[0]))
sns.scatterplot(x="Sales",y="PV",data = df, hue='PV')
```

```python
plt.title('UV vs Events(r = {0:0.2f})'.format(pearsonr(drop['UV'],drop['Events'])[0]))
sns.scatterplot(x="UV",y="Events",data = df, hue='Events')
```

```python
plt.title('UV vs Sales(r = {0:0.2f})'.format(pearsonr(drop['UV'],drop['Sales'])[0]))
sns.scatterplot(x="UV", y="Sales",hue="Events", style="Events", data=df)
plt.show()
```
